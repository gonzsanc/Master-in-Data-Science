```python
import numpy as np
import pandas as pd
from sklearn import preprocessing

trainPath = "../work_data/train.csv"
```

```python
dfTraincsv = pd.read_csv(trainPath)
```

```python
dfTraincsv.sample(5)
```

```python
#Normalizing the series...
salesMean = dfTraincsv["Sales"].mean()
salesStd = dfTraincsv["Sales"].std()
customersMean = dfTraincsv["Customers"].mean()
customersStd = dfTraincsv["Customers"].std()
salesMean, salesStd, customersMean, customersStd
```

```python
dfTraincsv["SalesNormalized"] = (dfTraincsv["Sales"]-salesMean)/salesStd
dfTraincsv["CustomersNormalized"] = (dfTraincsv["Customers"]-salesMean)/salesStd
dfTraincsv.sample(10)
```

```python
dfTraincsv["SalesPerCostumerNormalized"] = dfTraincsv["SalesNormalized"]/dfTraincsv["CustomersNormalized"]
dfTraincsv["SalesPerCostumer"] = dfTraincsv["Sales"]/dfTraincsv["Customers"]
```

```python
dfCorr = dfTraincsv.corr()
dfCorrConditioned = dfTraincsv.corr()[(dfTraincsv.corr()>0.5) & (dfTraincsv.corr()<1) ]
dfCorr
```
