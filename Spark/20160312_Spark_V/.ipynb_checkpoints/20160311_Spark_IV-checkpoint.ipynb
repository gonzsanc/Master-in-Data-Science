```python
f = gzip.GzipFile(dataPath)
df = pd.read_csv(f, sep="^")
```

```python
df.head()
```

```python
df.count()
```

```python
df.columns
```

```python
df = df[['bookings_seg', 'revenue_amount_seg', 'fuel_surcharge_amount_seg']]
```

```python
df.head()
```

```python
df['revenue'] = df['revenue_amount_seg'] / df['bookings_seg']
df['tax'] = df['fuel_surcharge_amount_seg'] / df['bookings_seg']
```

```python
df.head()
```

```python
simple = df[['revenue', 'tax']]
```

```python
simple.dtypes
```

```python
#smaller = simple.sample(10000)

plt.scatter(simple['revenue'], simple['tax'], alpha=0.1)

plt.show()
```

```python
plt.scatter(simple['revenue'], simple['tax'], alpha=0.1)

plt.xlim(-50, 2000)
plt.ylim(-50, 600)

plt.show()
```

```python
smaller = simple.sample(10000)
```

```python
plt.scatter(smaller['revenue'], smaller['tax'], alpha=0.1)

plt.xlim(-50, 2000)
plt.ylim(-50, 600)

plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
kmeans = KMeans(n_clusters=10)
```

```python
clusters = kmeans.fit_predict(smaller)
```

```python
clusters
```

```python
smaller.iloc[0]
```

```python
smaller['cluster'] = clusters
smaller.iloc[0]
```

```python
smaller.sample(5)
```

```python
groups = smaller.groupby(['cluster'])
```

```python
stdev= groups.std()
```

```python
groups.count()
```

```python
stdev
```

```python
plt.scatter(smaller['revenue'], \
            smaller['tax'], \
            c=smaller['cluster'], \
            alpha=0.5)

plt.xlim(-50, 2000)
plt.ylim(-50, 600)

plt.show()
```
